```python
import pandas as pd
import numpy as np
```

```python
df = (52.28749785526844, -113.77188576216403), (52.30275093297708, -113.8125427302129), (52.29330271059767, -113.84756164901398), (52.221559783492744, -113.81890823513834), (52.24458819060406, -113.76681127164842)
```

```python
df=pd.DataFrame(df, columns=["lat", "long"])
```

```python
df
```

## Plot Locations

```python
import folium
```

```python
df['lat'] = df['lat'].astype(float)
df['long'] = df['long'].astype(float)
```

```python
m = folium.Map(location=(df["lat"].mean(), df["long"].mean()), zoom_start=10, tiles="cartodb positron")
```

```python
for index, row in df.iterrows():
    marker = folium.CircleMarker(location=(row["lat"], row["long"]), radius=2)
    marker.add_to(m)
```

```python
m
```

## Warehouse Location if Equal Shipments

```python
marker2 = folium.CircleMarker(location=(df["lat"].mean(), df["long"].mean()), color='green', radius=4)
marker2.add_to(m)
```

```python
m
```

## Warehouse Location Based on Size of Shipment

```python
df["goods"] = [50, 7, 3, 42, 99]
```

```python
df
```

```python
weighted_avg_lat = np.average(df["lat"], weights=df["goods"])
```

```python
weighted_avg_lat
```

```python
weighted_avg_long = np.average(df["long"], weights=df["goods"])
```

```python
weighted_avg_long
```

```python
marker3 = folium.CircleMarker(location=(weighted_avg_lat, weighted_avg_long), color='red', radius=4)
marker3.add_to(m)
```
