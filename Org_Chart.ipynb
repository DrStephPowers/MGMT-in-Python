```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()

nodes = np.arange(0, 8).tolist()

G.add_nodes_from(nodes)

G.add_edges_from([(0,1), (0,2), (1,3), (1, 4), (2, 5), (2, 6), (2,7)])

pos = {0:(10, 10), 1:(7.5, 7.5), 2:(12.5, 7.5), 3:(6, 6), 4:(9, 6),
 5:(11, 6), 6:(14, 6), 7:(17, 6)}

labels = {0: "CEO", 1:"Team A Lead", 2: "Team B Lead", 3: "Staff A", 4: "Staff B",
 5: "Staff C", 6: "Staff D", 7: "Staff E"}

nx.draw_networkx(G, pos = pos, labels = labels, arrows = True,
 node_shape = "s", node_color = "white")

nx.draw_networkx_edge_labels(G, pos = pos, edge_labels={(0, 1): 'A', (0, 2): 'B'},
                             font_color='black')

plt.title("Org Chart")
plt.show()
```

## Degree Centrality

Percentage of nodes connected to

Percentage of people you interact with

```python
cen = nx.degree_centrality(G)
cen
```

```python
cen.values()
```

```python
plt.bar(["CEO", "Team A Lead", "Team B Lead", "Staff A", "Staff B",
 "Staff C", "Staff D", "Staff E"], cen.values())
plt.xticks(rotation=45)
plt.show()
```

## Eigenvector Centrality

Which node has the most influence

G adds the centrality of predecessors. G reverse adds the centrality of successors.

```python
eig = nx.eigenvector_centrality_numpy(G.reverse())
eig
```

```python
eig.values()
```

```python
plt.bar(["CEO", "Team A Lead", "Team B Lead", "Staff A", "Staff B",
 "Staff C", "Staff D", "Staff E"], eig.values())
plt.xticks(rotation=45)
plt.show()
```

## Betweeness Centrality

Times a node is part of the shortest path

People have to go through you to communicate in the office

```python
bet = nx.betweenness_centrality(G)
bet
```

```python
bet.values()
```

```python
plt.bar(["CEO", "Team A Lead", "Team B Lead", "Staff A", "Staff B",
 "Staff C", "Staff D", "Staff E"], bet.values())
plt.xticks(rotation=45)
plt.show()
```

## Closeness Centrality

Average length of path to other nodes

How many people do you have to go through to communicate with someone else

G is inward distance. G reverse is outward distance.

```python
clos = nx.closeness_centrality(G.reverse())
clos
```

```python
clos.values()
```

```python
plt.bar(["CEO", "Team A Lead", "Team B Lead", "Staff A", "Staff B",
 "Staff C", "Staff D", "Staff E"], clos.values())
plt.xticks(rotation=45)
plt.show()
```

### A more centralized organizational structure has less closeness and more betweeness for upper administration.
